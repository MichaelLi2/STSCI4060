In the following cell import the Python modules that are needed
for doing the KMeans clustering, 2D and 3D scatter plots. (4 points)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

In the cell below, from the data file, iris.txt (that can be downloaded from the coures website), create a list called dataset to hold the first three columns of the data points and another list called species to hold the values of the iris species names. Display the contents of these two lists. Convert the two lists to numpy arrays and keep the same names. (8 points)

```python
dataset = []
species = []

readin = open('./iris.txt')
for line in readin:
    splits = line.split(',')
    dataset.append([splits[0], splits[1], splits[2]])
    species.append(splits[4][:-1])
print 'Dataset:\n' + str(dataset)
print 'Species:\n' + str(species)

dataset = np.array(dataset)
species = np.array(species)
```

Convert the data type of dataset to float and display the shape of dataset. (2 points)

```python
dataset = dataset.astype(float)
dataset.shape
```

In the cell below, use the KMeans algorithm to do a K-Means cluster analysis using the first two columns of the data of the dataset array. The KMean settings are: n_cluster=3, init='random', n_init=10, max_iter=500, tol=1e-4, and random_state=0. Use the same settings for the rest of analyses below. Then, you create two scatter plots with a total figure size of 20X10: 1) a plot showing the K-Means clusting resutls and 2) a plot showing the actual clusters (represented by flower names). These two plots should be arranged in a 1X2 grid; the K-Means plot is diaplayed in position 1 and the actual cluster plot is in position 2. Use the same type of marker ('o'), set s=50, set labels to flower names, and set the colors as follows: red for Iris-setosa, blue for Iris-versicolor and orange for Iris-virginica. Note that the cluster names of the K-Means result are represented by numbers (0,1,2), but you should be able to figure out what number represents what kind of flower by comparing the numbers to the actual cluster names. Give the title "K-Means" to the K-Means plot and "Actual" to the actual cluster plot. The legends should be displayed in the upper right corner of each plot; the two plots should have the same legends. Make sure that you have the same color code patterns for both the clusters and the legends in the two plots (same below).  (29 points)

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=500, tol=1e-4, random_state=0)
X = dataset[:,:2]
y_km = km.fit_predict(X)
fig = plt.figure(figsize=(20,10))
fig1 = fig.add_subplot(121)
fig1.scatter(X[y_km==0,0], X[y_km==0,1], s= 50, c='red', marker='o', label='Iris-versicolor')
fig1.scatter(X[y_km==1,0], X[y_km==1,1], s= 50, c='blue', marker='o', label='Iris-setosa')
fig1.scatter(X[y_km==2,0], X[y_km==2,1], s= 50, c='orange', marker='o', label='Iris-virginica')
fig1.set_title('K-Means')
fig1.legend(loc=1)
fig1.grid()

fig2 = fig.add_subplot(122)
fig2.scatter(X[species=='Iris-setosa',0], X[species=='Iris-setosa',1], s= 50, c='blue', marker='o', label='Iris-setosa')
fig2.scatter(X[species=='Iris-versicolor',0], X[species=='Iris-versicolor',1], s= 50, c='red', marker='o', label='Iris-versicolor')
fig2.scatter(X[species=='Iris-virginica',0], X[species=='Iris-virginica',1], s= 50, c='orange', marker='o', label='Iris-virginica')
fig2.set_title('Actual')
fig2.legend(loc=1)
fig2.grid()
plt.show()

```

You may get a different clustering result when use a different set of data points. Use the same analysis with the same settings as the above analysis but use the second and third columns of the dataset array. Note that the cluster numbers of this K-Means analysis result may be different from the meaning of the result produced in the above analysis, so you may have to match the cluster numbers with the actual cluster names again. By looking at the scatter plots, you should see which group of the data points represents the actual situation better; add a cell below your resut of the following cell to write your comments. (29 points)

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=500, tol=1e-4, random_state=0)
X = dataset[:,1:]
y_km = km.fit_predict(X)
fig = plt.figure(figsize=(20,10))
fig1 = fig.add_subplot(121)
fig1.scatter(X[y_km==0,0], X[y_km==0,1], s= 50, c='orange', marker='o', label='Iris-virginica')
fig1.scatter(X[y_km==1,0], X[y_km==1,1], s= 50, c='red', marker='o', label='Iris-versicolor')
fig1.scatter(X[y_km==2,0], X[y_km==2,1], s= 50, c='blue', marker='o', label='Iris-setosa')
fig1.set_title('K-Means')
fig1.legend(loc=1)
fig1.grid()

fig2 = fig.add_subplot(122)
fig2.scatter(X[species=='Iris-setosa',0], X[species=='Iris-setosa',1], s= 50, c='blue', marker='o', label='Iris-setosa')
fig2.scatter(X[species=='Iris-versicolor',0], X[species=='Iris-versicolor',1], s= 50, c='red', marker='o', label='Iris-versicolor')
fig2.scatter(X[species=='Iris-virginica',0], X[species=='Iris-virginica',1], s= 50, c='orange', marker='o', label='Iris-virginica')
fig2.set_title('Actual')
fig2.legend(loc=1)
fig2.grid()
plt.show()
```

We can see that using the second two columns of the data gives better results than the first two.  The groups are more distinct and more separate on the graph, with less overlap between the Iris-versicolor and Iris-virginica in particular, leading to the k-means algorithm to correctly identify more of each group.

Now do your K-Means analysis again using all the three colums of dataset. In this case, you need to create a 3D scatter plot, using the same settings for plotting, except arranging the two plots in a 2X1 grid. (28 points)

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=500, tol=1e-4, random_state=0)
y_km = km.fit_predict(dataset)
fig = plt.figure(figsize=(10,20))
fig1 = fig.add_subplot(211, projection='3d')
fig1.scatter(dataset[y_km==0,0], dataset[y_km==0,1], dataset[y_km==0,2], color = 'orange', label='Iris-virginica')
fig1.scatter(dataset[y_km==1,0], dataset[y_km==1,1], dataset[y_km==1,2], color = 'red', label='Iris-versicolor')
fig1.scatter(dataset[y_km==2,0], dataset[y_km==2,1], dataset[y_km==2,2], color = 'blue', label='Iris-setosa')
fig1.set_title('K-Means')
fig1.legend(loc=1)
fig1.grid()

fig2 = fig.add_subplot(212, projection='3d')
fig2.scatter(dataset[species=='Iris-setosa',0], dataset[species=='Iris-setosa',1], dataset[species=='Iris-setosa',2], color = 'blue', label='Iris-setosa')
fig2.scatter(dataset[species=='Iris-versicolor',0], dataset[species=='Iris-versicolor',1], dataset[species=='Iris-versicolor',2], color = 'red', label='Iris-versicolor')
fig2.scatter(dataset[species=='Iris-virginica',0], dataset[species=='Iris-virginica',1], dataset[species=='Iris-virginica',2], color = 'orange', label='Iris-virginica')
fig2.set_title('Actual')
fig2.legend(loc=1)
fig2.grid()
plt.show()
```
